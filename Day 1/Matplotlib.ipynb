#### 1. Line Plot

#### Lets consider 2 list days and temperature

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,marker='+',color='red',markersize=15)
plt.xlabel(" Day ")
plt.ylabel("Temperature")
plt.title("Day wise temperature")
```

#### 2. Scatter Plot

Pandas is a software library written for the Python programming language for data manipulation and analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=pd.read_csv("Road_Accidents_2017-India.csv")
```

```python
data
```

```python
plt.scatter(data.Years,data['Total Number of Persons Killed (in numbers)'],marker=".")
plt.ylabel("No of people Died")
plt.xlabel("Year")
plt.title("People Died in Accident respect to Years in India")
plt.grid()
```

##### 3. Bar Chart

- bar chart is basically used for see the comparision between various parameters

```python
company=["TCS","Infy","LTI","PSL"]
revenue=[400,370,90,160]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(company,revenue,label="Revenue")
plt.xlabel("Company")
plt.ylabel("Revenue in Bill. Dollars")
plt.title("Graph for company vs Revenue")
plt.legend()
```

##### 4. Pie Charts

- it used to plot the graph with percentages for each criteria. for ex company expenses

```python
company_expenses=[100,60,40,20,80]
expense_data=["emp_sal","IF Devlopment","Food Expenses","Travelling Allow","prizes"]
```

```python
plt.pie(company_expenses,labels=expense_data,startangle=90, shadow=True,autopct='%1.2f%%',radius=1.5,explode=[0,0,0.1,0.2,0.2])
plt.savefig("TCE_expenses.png")
```
